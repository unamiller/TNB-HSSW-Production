# 3. Calculate polynya area from SIC 

```python
t=slice('2017-07-04 17:00','2017-07-04 17:05') #start and end ranges because instrument times are not exact
t2=slice('2017-07-05 11:00','2017-07-05 11:05')

SIC_threshold = 30 #%

#using 7/5 because satellite crosses TNB 4 am and 12pm, so majority of of this event occurs in window of 7/5 not 7/4

plt.scatter(SIC.lon,SIC.lat,c=SIC.ice_concentration)
plt.scatter(SIC.lon.where(SIC.ice_concentration<SIC_threshold),SIC.lat.where(SIC.ice_concentration<SIC_threshold),
            c='r')

cbar = plt.colorbar()
cbar.set_label('Sea Ice Concentration (%)')
#plt.scatter(163.96, -74.97, c='k', marker='^', s=200)
#plt.contourf(ds.lon,ds.lat,ds.ice_concentration, levels=[0,70])
plt.xlim([163,166])
plt.ylim([-74,-75.6])
plt.gca().invert_yaxis()
plt.ylabel('Lat')
plt.xlabel('Lon')
subtract = 24 #### Account for missing Nansen Ice Shelf area

p_area = 3.125**2 * (len(np.where(SIC.ice_concentration.values < SIC_threshold)[0]) + subtract)

avg_SIC = np.where(SIC.ice_concentration.values < SIC_threshold)[0].mean()
```

# 4. Calculate HSSW Production rate

Following Equations 1-4 in the manuscript.

```python
tslice=slice(t.start, t2.stop) 

psalmin = mooring.practical_salinity.mean() -  mooring.practical_salinity.std()   
s47 = mooring.practical_salinity.sel(time=tslice)
rho47 = gsw.density.rho(gsw.SA_from_SP(mooring.practical_salinity,47, 163.96,-74.97), mooring.conservative_temperature,47).sel(time=tslice)

###### Equation 1 #####
m_s = (rho47.sel(time=tslice)*s47.sel(time=tslice).values)*(30-0) # Equation 1


linear_model=np.polyfit(range(len(m_s.dropna(dim='time').values)),m_s.dropna(dim='time'),1)
###### Equation 2 #####
dm_dt1 = linear_model[0]/60 #convert from minutes to seconds  
###### Equation 3 #####
P_s = dm_dt1 * p_area * 10**6 #convert km^2 to m^2  

psalmin = 34.7
linear_model_sal=np.polyfit(range(len(s47.where(np.abs(np.diff(s47, prepend = [100]))<0.01).dropna(dim='time').values)), # Remove fresh spikes in salinity time series (Caused by frazil ice contamination)
                            s47.where(np.abs(np.diff(s47, prepend = [100]))<0.01).dropna(dim='time'),1)
x_sal = range(len(s47.values))
y_sal = linear_model_sal[0]*x_sal + linear_model_sal[1]

linear_model_rho=np.polyfit(range(len(rho47.dropna(dim='time').values)),rho47.dropna(dim='time'),1)
x_rho = range(len(rho47.values))
y_rho = linear_model_rho[0]*x_rho + linear_model_rho[1]

###### Equation 4 #####
P_hssw_linear = P_s/(y_rho[-1]*(y_sal[-1]-y_sal[0]))* 10**-6


plt.plot(x_sal,y_sal)
plt.plot(s47.where(s47 > psalmin).dropna(dim='time'), label = 'Original')
plt.plot(s47.where(np.abs(np.diff(s47, prepend = [100]))<0.01).dropna(dim='time'), label = 'Despiked')
plt.title(str(tslice))
plt.ylabel('Practical Salinity')
plt.legend()

#prior 1 week slice
slice1week = slice(str(pd.to_datetime(t.start) - pd.Timedelta(days=7)),t.stop)
slice1week_wind = aws.sel(time=slice1week).wspd
#     avg_U_1week_g25 = aws.sel(time=slice1week).where(aws.wspd > 25).wspd.mean() #avg wind speed >25
#     avg_U_1week = aws.sel(time=slice1week).wspd.mean() #avg wind speed
#     count_U_1week = aws.sel(time=slice1week).where(aws.wspd > 25).wspd.count()/6 #number of hours of winds above 25ms


currents_mean = np.abs(-np.cos(0.785398)*currents.east[:14]).mean().values
currents_std = np.abs(-np.cos(0.785398)*currents.east[:14]).std().values
current_profile = np.abs(-np.cos(0.785398)*currents.east[:14]).mean(axis=1).values

df_HSSW = pd.DataFrame([[P_hssw_linear,dm_dt1,t2.stop,y_sal[0], y_sal[-1],y_rho[-1], p_area, avg_SIC, slice1week_wind.values,                                                
                         ]],
                  columns=['P_hssw','dm_dt','tstop','sal_start', 'sal_end', 'den_end','p_area', 'avg_SIC', 'slice1week'],
                  index=[t.start])
```
