# 3. Calculate polynya area from SIC 

```python
t=slice('2017-07-04 17:00','2017-07-04 17:05') #start and end ranges because instrument times are not exact
t2=slice('2017-07-05 11:00','2017-07-05 11:05')

SIC_threshold = 30 #%

#using 7/5 because satellite crosses TNB 4 am and 12pm, so majority of of this event occurs in window of 7/5 not 7/4

plt.scatter(SIC.lon,SIC.lat,c=SIC.ice_concentration)
plt.scatter(SIC.lon.where(SIC.ice_concentration<SIC_threshold),SIC.lat.where(SIC.ice_concentration<SIC_threshold),
            c='r')

cbar = plt.colorbar()
cbar.set_label('Sea Ice Concentration (%)')
#plt.scatter(163.96, -74.97, c='k', marker='^', s=200)
#plt.contourf(ds.lon,ds.lat,ds.ice_concentration, levels=[0,70])
plt.xlim([163,166])
plt.ylim([-74,-75.6])
plt.gca().invert_yaxis()
plt.ylabel('Lat')
plt.xlabel('Lon')
NIS_correction = 24 #### Account for missing Nansen Ice Shelf area

p_area = 3.125**2 * (len(np.where(SIC.ice_concentration.values < SIC_threshold)[0]) + NIS_correction)

avg_SIC = np.where(SIC.ice_concentration.values < SIC_threshold)[0].mean()
```

# 4. Calculate HSSW Production rate

Following Equations 1-4 in the manuscript.

```python
tslice=slice(t.start, t2.stop) 

sbe = mooring

psalmin = mooring.practical_salinity.mean() -  mooring.practical_salinity.std()   
s_mean = np.mean([sbe.practical_salinity.sel(depth=47).sel(time=tslice).dropna(dim='time').values, 
                sbe.practical_salinity.sel(depth=119).sel(time=tslice).dropna(dim='time').values,
                sbe.practical_salinity.sel(depth=144).sel(time=tslice).dropna(dim='time').values,
                sbe.practical_salinity.sel(depth=168).sel(time=tslice).dropna(dim='time').values,
                sbe.practical_salinity.sel(depth=219).sel(time=tslice).dropna(dim='time').values,
                sbe.practical_salinity.sel(depth=360).sel(time=tslice).dropna(dim='time').values], axis = 0)
    
rho_mean = np.mean([sbe.sigma0.sel(depth=47).sel(time=tslice).dropna(dim='time').values+1000, 
                sbe.sigma0.sel(depth=119).sel(time=tslice).dropna(dim='time').values+1000,
                sbe.sigma0.sel(depth=144).sel(time=tslice).dropna(dim='time').values+1000,
                sbe.sigma0.sel(depth=168).sel(time=tslice).dropna(dim='time').values+1000,
                sbe.sigma0.sel(depth=219).sel(time=tslice).dropna(dim='time').values+1000,
                sbe.sigma0.sel(depth=360).sel(time=tslice).dropna(dim='time').values+1000], axis = 0)
sigma_diff = np.mean(sbe.sigma0.sel(depth=360).sel(time=tslice).dropna(dim='time').values) - np.mean(sbe.sigma0.sel(depth=47).sel(time=tslice).dropna(dim='time').values)
        
###### Equation 1 #####
m_s = np.trapz([(sbe.sigma0.sel(depth=47).sel(time=tslice).dropna(dim='time').values+1000)*sbe.practical_salinity.sel(depth=47).sel(time=tslice).dropna(dim='time').values, 
            (sbe.sigma0.sel(depth=119).sel(time=tslice).dropna(dim='time').values+1000)*sbe.practical_salinity.sel(depth=119).sel(time=tslice).dropna(dim='time').values,
            (sbe.sigma0.sel(depth=144).sel(time=tslice).dropna(dim='time').values+1000)*sbe.practical_salinity.sel(depth=144).sel(time=tslice).dropna(dim='time').values,
            (sbe.sigma0.sel(depth=168).sel(time=tslice).dropna(dim='time').values+1000)*sbe.practical_salinity.sel(depth=168).sel(time=tslice).dropna(dim='time').values,
            (sbe.sigma0.sel(depth=219).sel(time=tslice).dropna(dim='time').values+1000)*sbe.practical_salinity.sel(depth=219).sel(time=tslice).dropna(dim='time').values,
            (sbe.sigma0.sel(depth=360).sel(time=tslice).dropna(dim='time').values+1000)*sbe.practical_salinity.sel(depth=360).sel(time=tslice).dropna(dim='time').values], 
                   x = np.array([47,119,144,168,219,360]), axis = 0)
              

linear_model=np.polyfit(range(len(m_s)),m_s,1)
###### Equation 2 #####
dm_dt1 = linear_model[0]/60 #convert from minutes to seconds  
###### Equation 3 #####
P_s = dm_dt1 * p_area * 10**6 #convert km^2 to m^2  

psalmin = 34.7
linear_model_sal=np.polyfit(range(len(s_mean[np.where(np.abs(np.diff(s_mean, prepend = [100]))<0.01)])),
                                s_mean[np.where(np.abs(np.diff(s_mean, prepend = [100]))<0.01)],1)
x_sal = range(len(s_mean))
y_sal = linear_model_sal[0]*x_sal + linear_model_sal[1]

linear_model_rho=np.polyfit(range(len(rho_mean[np.where(np.abs(np.diff(rho_mean, prepend = [100]))<0.01)])),
                            rho_mean[np.where(np.abs(np.diff(rho_mean, prepend = [100]))<0.01)],1)
x_rho = range(len(rho_mean))
y_rho = linear_model_rho[0]*x_rho + linear_model_rho[1]

###### Equation 4 #####
P_hssw_linear = P_s/(y_rho[-1]*(y_sal[-1]-y_sal[0]))* 10**-6


plt.plot(range(len(m_s)),m_s)
plt.plot(range(len(m_s)),range(len(m_s))*linear_model[0] + linear_model[1])

#prior 1 week slice
slice1week = slice(str(pd.to_datetime(t.start) - pd.Timedelta(days=7)),t.stop)
slice1week_wind = aws.sel(time=slice1week).wspd
#     avg_U_1week_g25 = aws.sel(time=slice1week).where(aws.wspd > 25).wspd.mean() #avg wind speed >25
#     avg_U_1week = aws.sel(time=slice1week).wspd.mean() #avg wind speed
#     count_U_1week = aws.sel(time=slice1week).where(aws.wspd > 25).wspd.count()/6 #number of hours of winds above 25ms


currents_mean = np.abs(-np.cos(0.785398)*currents.east[:14]).mean().values
currents_std = np.abs(-np.cos(0.785398)*currents.east[:14]).std().values
current_profile = np.abs(-np.cos(0.785398)*currents.east[:14]).mean(axis=1).values

df_HSSW = pd.DataFrame([[P_hssw_linear,dm_dt1,t2.stop,y_sal[0], y_sal[-1],y_rho[-1], p_area, avg_SIC, slice1week_wind.values,                                                
                         ]],
                  columns=['P_hssw','dm_dt','tstop','sal_start', 'sal_end', 'den_end','p_area', 'avg_SIC', 'slice1week'],
                  index=[t.start])
```
